```python
import numpy as np
import matplotlib.pyplot as plt
#plt.style.use('./deeplearning.mplstyle')
```

```python
x_train = np.array([1.0, 2.0])
y_train = np.array([300.0, 500.0])
print(f"The input feature is: {x_train}")
print(f"The output target is: {y_train}")
print(x_train.shape)
```

```python
print(f"The shape of training data: {x_train.shape[0]}")
m = x_train.shape[0]
print(f"The number of training examples: {m}")
```

```python
plt.scatter(x_train, y_train, marker = "x", c = "r", label = "given data")
plt.title("Price vs Size")
plt.ylabel("Prices")
plt.xlabel("Size")
plt.legend()
plt.show()
```

```python
def compute_model_output(w, b, x):
    m = x.shape[0]
    output = np.zeros(m)
    for i in range(m):
        output[i] = w * x[i] + b
    return output
```

```python
y_pred = compute_model_output(100, 100, x_train)
plt.plot(x_train, y_pred, marker = "x", color = "b", label = "prediction")
plt.scatter(x_train, y_train, marker = "x", color = "r", label = "given data")
```
